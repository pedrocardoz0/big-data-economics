```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import time
from google.colab import userdata
from fredapi import Fred

plt.style.use('fivethirtyeight')
pd.set_option('display.max_columns', 500)
color_pallet = plt.rcParams["axes.prop_cycle"].by_key()["color"]

fred_key = userdata.get('fre-api')
```

```python
fred = Fred(api_key=fred_key)
```

## Busca pelas taxas de desemprego de outros paises.


```python
unemp_df = fred.search('unemployment rate', filter=('frequency', 'Annual'))
unemp_df.query('units == "Percent"')
unemp_df = unemp_df.loc[unemp_df['title'].str.contains('Unemployment Rate: Aged 15 and over: All Persons for ')]
unemp_df = unemp_df.loc[unemp_df['seasonal_adjustment_short'].str.contains('N')]
unemp_df
```

```python
all_datasets = []

for myid in unemp_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_datasets.append(results)
    time.sleep(0.1)


uemp_results = pd.concat(all_datasets, axis=1)
```

## Ajuste no nome da coluna de cada pais

```python
uemp_copy = uemp_results.copy()

id_to_country = unemp_df['title'].str.replace('Unemployment Rate: Aged 15 and over: All Persons for ','').to_dict()
uemp_copy.columns = [id_to_country[c] for c in uemp_copy.columns]
```


## VisualizaÃ§Ã£o da taxa de desemprego de cada pais

```python
px.line(uemp_copy)
```

## Busca pelo gasto pÃºblico de cada paÃ­s.

```python
inflation_df = fred.search('inflation rate nation', filter=('frequency', 'Annual'))
inflation_df.query('units == "Percent"')
inflation_df = inflation_df.loc[inflation_df['title'].str.contains('Consumer Price Index: OECD Groups: All Items Non-Food Non-Energy: Total for ')]
inflation_df
```

```python
all_inflation_datasets = []

for myid in inflation_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_inflation_datasets.append(results)
    time.sleep(0.1)


inflation_results = pd.concat(all_inflation_datasets, axis=1)
```

## Ajuste no nome da coluna de cada pais.

```python
inflation_copy = inflation_results.copy()

id_to_country = inflation_df['title'].str.replace('Consumer Price Index: OECD Groups: All Items Non-Food Non-Energy: Total for ','').to_dict()
inflation_df = inflation_df.loc[inflation_df['seasonal_adjustment_short'].str.contains('N')]
inflation_copy.columns = [id_to_country[c] for c in inflation_copy.columns]
inflation_copy = inflation_copy.loc[:, ~inflation_copy.columns.duplicated()]
```

## Visualizando a inflaÃ§Ã£o de cada pais

```python
px.line(inflation_copy)
```

## Visualizando a InflaÃ§Ã£o e Taxa de Desemprego.

```python
country = 'United States'
fig, ax = plt.subplots(figsize=(10, 5), sharex=True)
ax2 = ax.twinx()
uemp_copy_2, inflation_copy_2 = uemp_copy, inflation_copy
l1 = uemp_copy_2[country] \
    .plot(ax=ax, label='Unemployment')
l2 = inflation_copy_2[country] \
    .plot(ax=ax2, label='Inflation', color=color_pallet[1])

ax2.grid(False)
ax.set_title(country)
fig.legend(labels=['Unemployment', 'Inflation'])
plt.show()
```

```python
fig, axs = plt.subplots(10, 5, figsize=(30, 30), sharex=True)
axs = axs.flatten()

i = 0
for country in uemp_copy_2.columns:
    if country in ["Brazil", "Indonesia", "Russia"]:
      continue
    ax2 = axs[i].twinx()
    uemp_copy_2[country] \
        .plot(ax=axs[i], label='Unemployment')
    inflation_copy_2[country] \
        .plot(ax=ax2, label='Inflation', color=color_pallet[1])
    ax2.grid(False)
    axs[i].set_title(country)
    i += 1
plt.tight_layout()
plt.show()
```
